```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
music_df = pd.read_csv('df_audio_features_10', index_col='song_name')
music_df.drop(['id', 'html', 'artist'], axis=1, inplace=True)
music_df.head()
```

# Min Max Scaler

```python
# Import MinMax Scaler
from sklearn.preprocessing import MinMaxScaler

# Create DataFrame of MinMax scaled values
music_min_max_df = MinMaxScaler().fit_transform(music_df)
music_minmax = pd.DataFrame(music_min_max_df, columns=music_df.columns, index=music_df.index)

# Display x column of original DataFrame and x column of MinMaxed DataFrame to compare
pd.DataFrame({
    'original': music_df.iloc[:,1], 
    'min_max_scale': music_minmax.iloc[:,1]
}).sort_values(by='original')
```

# Standard Scaler

```python
# Import Standard Scaler
from sklearn.preprocessing import StandardScaler

# Create DataFrame of Standar scaled values
music_standard_df = StandardScaler().fit_transform(music_df)
music_standard = pd.DataFrame(music_standard_df, columns=music_df.columns, index=music_df.index)

# Display x column of original DataFrame and x column of MinMaStandard scaledxed DataFrame to compare
pd.DataFrame({
    'original': music_df.iloc[:,1], 
    'standard_scale': music_standard.iloc[:,1]
}).sort_values(by='original')
```

# Robost Scaler

```python
# Import Robust Scaler
from sklearn.preprocessing import RobustScaler

# Create DataFrame of Robust scaled values
music_robust_df = RobustScaler().fit_transform(music_df)
music_robust = pd.DataFrame(music_robust_df, columns=music_df.columns, index=music_df.index)

# Display x column of original DataFrame and x column of Robust DataFrame to compare
pd.DataFrame({
    'original': music_df.iloc[:,1], 
    'robust_scale': music_robust.iloc[:,1]
}).sort_values(by='original')
```

# QuantileTransformer

```python
# Import Quantile Transformer
from sklearn.preprocessing import QuantileTransformer

# Create DataFrame of Quantile transformed values
music_quantile_df = QuantileTransformer(n_quantiles=len(music_df.iloc[:,1])).fit_transform(music_df)
music_quantile = pd.DataFrame(music_quantile_df, columns=music_df.columns, index=music_df.index)

# Display x column of original DataFrame and x column of Quantile DataFrame to compare
pd.DataFrame({
    'original': music_df.iloc[:,1], 
    'quantile_transformed': music_quantile.iloc[:,1]
}).sort_values(by='original')
```

# Power Transformer

```python
# Import Power Transformer
from sklearn.preprocessing import PowerTransformer

# Create DataFrame of Power transformed values
music_power_df = PowerTransformer().fit_transform(music_df)
music_power = pd.DataFrame(music_power_df, columns=music_df.columns, index=music_df.index)

# Display 1st column of original DataFrame and 1st column of Power transformed DataFrame to compare
pd.DataFrame({
    'original': music_df.iloc[:,1], 
    'power_transformed': music_power.iloc[:,1]
}).sort_values(by='original')
```

# SciKit Learn - Euclidean Distance

```python
from sklearn.metrics import pairwise_distances

pd.DataFrame(pairwise_distances(music_df), index=music_df.index, columns=music_df.index)
```
